```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Load iris.csv into pandas dataframe
Heart_disease = pd.read_csv('C:/Austine/Blog/ML/4. Histogran/Age_Heart Disease.csv')
```

```python
Heart_disease.columns
```

```python
Sufferers = Heart_disease.loc[Heart_disease['Heart Disease'] == 1];
Non_sufferers = Heart_disease.loc[Heart_disease['Heart Disease'] == 0];
```

```python
#People suffering from heart disease
plt.hist(Sufferers['Age'])
plt.xlabel('Age of people suffering from Heart Disease')
plt.savefig('Age of people suffering from Heart Disease.png')
plt.show()
```

```python
#Histogram
sns.FacetGrid(Heart_disease, hue = 'Heart Disease', size = 5) \
    .map(sns.distplot, 'Age', kde=True) \
    .add_legend();
#plt.savefig('Histogram with Kernel Density Plot.png')
plt.show()
```
